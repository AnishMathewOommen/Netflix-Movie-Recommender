```python
# Read data from a CSV file that is inside a ZIP archive
# - 'header=None' indicates that the CSV file does not contain a header row (column names)
# - 'names=['Cust_Id', 'Rating']' specifies the column names for the resulting DataFrame
# - 'usecols=[0, 1]' specifies to read only the first two columns from the CSV file

netflix_dataset = pd.read_csv(
    'dataset/netflix_dataset.zip',  
    header=None,  
    names=['Cust_Id', 'Rating'],  
    usecols=[0, 1]
)

netflix_dataset.head(5)
```

```python
# Get the count of missing values (NaN) for each column in the DataFrame
missing_values_count = netflix_dataset.isnull().sum()

# Extract the count of missing values specifically from the 'Rating' column
missing_ratings_count = missing_values_count["Rating"]

# Output the count of missing values in the 'Rating' column
missing_ratings_count
```

```python
# Calculate the number of unique customers in the 'Cust_Id' column of the 'netflix_dataset' DataFrame
customer_count = netflix_dataset['Cust_Id'].nunique()

# Output the unique customer count
customer_count
```

```python
# Calculate the total number of ratings in the 'Cust_Id' column of the 'netflix_dataset' DataFrame
# 'count()' counts all non-NaN values in the 'Cust_Id' column
# 'movie_count' represents the total number of movies in the dataset

rating_count = netflix_dataset['Cust_Id'].count() - missing_ratings_count

# Output the total rating count
rating_count
```

```python
# Group the 'netflix_dataset' DataFrame by the 'Rating' column
# For each unique rating value, calculate the count of occurrences of that rating
rating_value = netflix_dataset.groupby('Rating')['Rating'].agg(['count'])

# Output the DataFrame with counts of each rating value
rating_value
```

```python
# Plot the horizontal bar chart
ax = rating_value.plot(kind='barh', legend=False, figsize=(15,10), edgecolor='black')

# Set the title with proper description and formatting
plt.title('Distribution of Customer Ratings\n'
          f'Total Movies: {missing_ratings_count}\n'
          f'Total Customers: {customer_count}\n'
          f'Total Ratings Given: {rating_count}', 
          fontsize=22, fontweight='bold')

# Label the x-axis and y-axis
plt.xlabel('Number of Ratings', fontsize=16)
plt.ylabel('Rating Value', fontsize=16)

# Add grid lines for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Customize the tick parameters for better visibility
plt.tick_params(axis='both', labelsize=16)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Initialize variables
current_movie_id = None  # To hold the current movie ID during processing
movie_id_list = []       # List to store movie IDs

# Iterate over each customer ID in the 'Cust_Id' column of the 'netflix_dataset' DataFrame
for customer_id in netflix_dataset['Cust_Id']:
    # Check if the customer ID contains a colon
    if ':' in customer_id:
        # Extract the numeric movie ID by removing the colon and convert it to an integer
        current_movie_id = int(customer_id.replace(':', ''))
    # Append the current movie ID to the list
    movie_id_list.append(current_movie_id)

# Output the list of movie IDs
movie_id_list
```

```python
# Add the new column to the Netflix dataset
netflix_dataset['Movie_Id'] = movie_id_list

# Removing the NAN Rows
netflix_dataset = netflix_dataset[netflix_dataset['Rating'].notna()]
netflix_dataset.head(5)
```

```python
netflix_dataset.info()
```

```python
# Changing the Cust_Id column datatype from object to integer
netflix_dataset['Cust_Id'] = netflix_dataset['Cust_Id'].astype('int')
netflix_dataset.info()
```

```python
# Pre-Filtering of Data
# Remove all the users who have rated only few movies
# Remove the movies that has been rated less in numbers

# Group the dataset by 'Movie_Id' and count the number of ratings for each movie
# This step summarizes how many ratings each movie has received
movie_rating_summary = netflix_dataset.groupby('Movie_Id')['Rating'].agg(['count'])

# Output the summary of ratings for each movie
movie_rating_summary
```

```python

# Calculate the 60th percentile of the number of ratings for each movie
# This percentile will be used as a benchmark to filter out movies with low rating counts
rating_threshold = round(movie_rating_summary['count'].quantile(0.6), 0)

# Output the rating threshold value
rating_threshold
```

```python
# Identify movies that have fewer ratings than the threshold
# Filter the movie_rating_summary DataFrame to find movies with rating counts below the rating_threshold
# Extract the indices (movie IDs) of these movies to create a list of movies to drop
movies_to_drop = movie_rating_summary[movie_rating_summary['count'] < rating_threshold].index

# Output the list of movie IDs to be dropped
movies_to_drop
```

```python
# Length of movies that are going to be dropped
len(movies_to_drop)
```

```python
customer_summary = netflix_dataset.groupby('Cust_Id')['Rating'].agg(['count'])
customer_summary
```

```python
# Calculate the 60th percentile of the number of ratings given by each customer
# This percentile will be used as a benchmark to filter out customers who have rated fewer movies
customer_rating_threshold = round(customer_summary['count'].quantile(0.6), 0)

# Output the customer rating threshold value
customer_rating_threshold
```

```python
# Identify customers who have rated fewer movies than the customer_rating_threshold
# Filter the customer_summary DataFrame to find customers with rating counts below the threshold
# Extract the indices (customer IDs) of these customers to create a list of customers to drop
inactive_customers = customer_summary[customer_summary['count'] < customer_rating_threshold].index

# Output the list of customer IDs to be dropped
inactive_customers
```

```python
# Length of In-active customers
len(inactive_customers)
```

```python
# Remove rows where 'Movie_Id' is in the drop_movie_list
netflix_dataset = netflix_dataset[~netflix_dataset['Movie_Id'].isin(movies_to_drop)]

# Remove rows where 'Cust_Id' is in the drop_cust_list
netflix_dataset = netflix_dataset[~netflix_dataset['Cust_Id'].isin(inactive_customers)]

# Print the shape of the DataFrame after filtering
print('After the trimming, the shape is: {}'.format(netflix_dataset.shape))
```

# Model Building

```python
# Load movie titles and related information from a CSV file into a DataFrame
# The CSV file is located at '/home/anish/Documents/netflix_movie_titles.csv'
# Specify the encoding as 'ISO-8859-1' to handle special characters
# Read only the first three columns (0, 1, and 2) from the CSV file
# Rename the columns to 'Movie_Id', 'Year', and 'Movie_name'
netflix_movie_titles = pd.read_csv('dataset/netflix_movie_titles.csv', 
                                    encoding='ISO-8859-1', 
                                    header=None, 
                                    usecols=[0, 1, 2], 
                                    names=['Movie_Id', 'Year', 'Movie_name'])

# Output the DataFrame containing movie titles and related information
netflix_movie_titles
```

```python

from surprise import Reader, Dataset, SVD
from surprise.model_selection import cross_validate
```

```python
# Define the format of the data for the Surprise library
# Specify that the data consists of 'Cust_Id', 'Movie_Id', and 'Rating' columns
reader = Reader(rating_scale=(1, 5))  # Adjust the rating scale according to your dataset

# Load a subset of the data (first 100,000 rows) into a Dataset object
# This is needed for the Surprise library to perform model training and evaluation
data = Dataset.load_from_df(netflix_dataset[['Cust_Id', 'Movie_Id', 'Rating']][:100000], reader)

# Initialize the SVD (Singular Value Decomposition) model for collaborative filtering
svd_model = SVD()

# Perform cross-validation to evaluate the SVD model's performance
# Use RMSE (Root Mean Squared Error) as the evaluation metric
# Perform 3-fold cross-validation
results = cross_validate(svd_model, data, measures=['RMSE'], cv=3, verbose=True)

```

```python
# We are interested in user with ID 2632461
# Create a copy of the DataFrame 'netflix_movie_titles' and assign it to the variable 'user_2632461'
user_2632461 = netflix_movie_titles.copy()

# Output the DataFrame for user 2632461
user_2632461
```

```python
user_2632461=user_2632461[~user_2632461['Movie_Id'].isin(movies_to_drop)]
user_2632461
```

```python
# Add a new column 'Estimate_Score' to the DataFrame 'user_2632461'
# This column contains estimated ratings for movies based on the SVD model
# Use the SVD model to predict ratings for each movie ID in the DataFrame
# The lambda function applies the 'predict' method of the SVD model to estimate the score for user 2632461 and each movie ID
user_2632461['Estimate_Score'] = user_2632461['Movie_Id'].apply(lambda x: svd_model.predict(2632461, x).est)

# Output the updated DataFrame with the estimated scores
user_2632461
```

```python
# Sort the DataFrame 'user_2632461' by the 'Estimate_Score' column in descending order
# This will reorder the movies based on their estimated scores for user 2632461, with the highest scores first
sorted_user_2632461 = user_2632461.sort_values('Estimate_Score', ascending=False)

# Output the sorted DataFrame
sorted_user_2632461
```

```python
# Sort the DataFrame 'user_2632461' by the 'Estimate_Score' column in descending order
# Select the top 5 movies with the highest estimated scores for user 2632461
top5_movies = user_2632461.sort_values('Estimate_Score', ascending=False).head(5)

# Output the DataFrame containing the top 5 movies
top5_movies
```
